## 37.i Create DataFrame with patients over 50 years old
## 37.ii Create the plot
## 37.iii Plot the data
## 37.iv Customize the plot
## 37.v Add a meanline
Beatiful, now you've created a plot of two different variables, let's change the style.

```python
over50 = heart_disease[heart_disease["age"] > 50]

fig, ax = plt.subplots()

scatter = ax.scatter(over50["age"], over50["chol"], c=over50["target"])

ax.set(
    title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol"
)
ax.legend(*scatter.legend_elements(), title = "Target")

ax.axhline(over50["chol"].mean(), linestyle = "--");
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

```python
fig, ax = plt.subplots()

scatter = ax.scatter(over50["age"], over50["chol"], c=over50["target"])

ax.set(
    title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol"
)
ax.legend(*scatter.legend_elements(), title = "Target")

ax.axhline(over50["chol"].mean(), linestyle = "--");
```

## 41.i Replot the same figure as above except change the `cmap` parameter of `scatter()` to `winter`

## 41.iii Create the plot
## 41.iv Plot the data

Beautiful! Now our figure has an upgraded color scheme let's save it to file.

```python
fig, ax = plt.subplots()

scatter = ax.scatter(over50["age"], over50["chol"], c=over50["target"], cmap="winter")

ax.set(
    title="Heart Disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol"
)
ax.legend(*scatter.legend_elements(), title = "Target")

ax.axhline(over50["chol"].mean(), linestyle = "--", color = "red");
```
